```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
generated_images = np.load("generated_images.npy")
generation = len(generated_images)-1

for image_index, image in enumerate(generated_images[generation]):
    print("Generation "+str(generation)+" image "+str(image_index))
    plt.imshow(np.array(image).reshape((28,28)))
    plt.show()
```
