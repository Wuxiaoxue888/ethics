# Find closest looking real image

```python
real_images = pd.read_csv("real_images.csv")
def find_closest_looking_real_image(image):
    closest_looking_image_index = np.argmin( np.count_nonzero(real_images.to_numpy() - image, axis=1) )
    return closest_looking_image_index, real_images.iloc[closest_looking_image_index].values.reshape((28,28))
```

```python
image = generated_images[10]
index, closest_looking_image = find_closest_looking_real_image(image)
plt.imshow(image.reshape((28, 28)), cmap='Greys')
plt.show()
plt.savefig('image_1.png')
print("Closest looking image index", index)
plt.imshow(closest_looking_image, cmap='Greys')
plt.show()
```
