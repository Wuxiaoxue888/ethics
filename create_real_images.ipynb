```python
# Import
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from datasets import load_dataset
```

```python
# Load the dataset. Uses streaming which means that the whole dataset is not loaded (its too big, 50 million images), instead one iterates through the dataset
dataset = load_dataset("quickdraw", split='train', streaming=True)
dataset
```

```python
# Example of how to iterate through the dataset, getting the next image
instance = next(iter(dataset))
instance
```

```python
# View the image
instance["image"]
```

```python
# Shuffle the dataset
shuffled_dataset = dataset.shuffle(seed=42, buffer_size=5_000_000)
```

```python
# Take the first 25 000 instances from the shuffled dataset
dataset_head = shuffled_dataset.take(25_000)
```

```python
# Turn the dataset into a list
# Warning! Takes 10 min to run on my computer
dataset_list = list(dataset_head)
```

```python
# Get images and turn each one into flattened np array (array with 784=28x28 int values)
images = list((map(lambda dict: np.array(dict["image"]).flatten(), dataset_list)))
```

```python
# Create dataframe from the images
# Warning! Takes 10 min to run on my computer
df = pd.DataFrame(data=images)
df
```

```python
# Save the DataFrame
df.to_csv('real_images.csv', index=False)
```
