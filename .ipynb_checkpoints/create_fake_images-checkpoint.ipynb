```python
# Import
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Function that creates a random image. Returns a 784=28*28 long numpy array of integers between 0 and 255
def create_random_image():
    return np.random.randint(256, size=28*28)
```

```python
# Create 25 000 random images
random_images = [create_random_image() for _ in range(25_000)]
```

```python
# Create dataframe from the images
# Warning! Takes 10 min to run on my computer
df = pd.DataFrame(data=random_images)
df
```

```python
# Save the DataFrame
df.to_csv('fake_images.csv', index=False)
```

# Show an image

```python
image = df.iloc[4].to_numpy().reshape((28, 28))
plt.imshow(image, interpolation='nearest')
plt.show()
```
