```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
generated_images = np.load("generated_images.npy")
```

```python
for generation_index, generation in enumerate(generated_images):
    for image_index, image in enumerate(generation):
        print("Generation "+str(generation_index)+" image "+str(image_index))
        plt.imshow(np.array(image).reshape((28,28)))
        plt.show()
```
